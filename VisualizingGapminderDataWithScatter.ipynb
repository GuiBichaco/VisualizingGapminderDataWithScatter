#2. Criando o gráfico Scatter

(Creating the Scatter Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    'country': ['Country A', 'Country B', 'Country C', 'Country D', 'Country E'],
    '1985': [100, 200, 150, 300, 250],
    '1986': [110, 210, 160, 310, 260],
    '1987': [120, 220, 170, 320, 270],
    '1988': [130, 230, 180, 330, 280],
    '1989': [140, 240, 190, 340, 290],
    '1990': [150, 250, 200, 350, 300],
    '1991': [160, 260, 210, 360, 310],
    '1992': [170, 270, 220, 370, 320],
    '1993': [180, 280, 230, 380, 330],
    '1994': [190, 290, 240, 390, 340],
    '1995': [200, 300, 250, 400, 350],
    '1996': [210, 310, 260, 410, 360],
    '1997': [220, 320, 270, 420, 370],
    '1998': [230, 330, 280, 430, 380],
    '1999': [240, 340, 290, 440, 390],
    '2000': [250, 350, 300, 450, 400],
    '2001': [260, 360, 310, 460, 410],
    '2002': [270, 370, 320, 470, 420],
    '2003': [280, 380, 330, 480, 430],
    '2004': [290, 390, 340, 490, 440],
    '2005': [300, 400, 350, 500, 450],
    '2006': [310, 410, 360, 510, 460],
    '2007': [320, 420, 370, 520, 470],
    '2008': [330, 430, 380, 530, 480]
}

# Converter o dicionário em DataFrame
df = pd.DataFrame(data)

# Selecionar os dados para o scatter plot
x = df['1985']
y = df['2008']
size = df['1995']  # Tamanho da bolinha
color = df['2005']  # Cor

# Criar o scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=size, c=color, alpha=0.7, cmap='viridis')
plt.title('Número Total de Mortes no Trânsito por País (1985 vs 2008)')
plt.xlabel('Número de Mortes no Trânsito em 1985')
plt.ylabel('Número de Mortes no Trânsito em 2008')
plt.colorbar(label='Variação de Mortes no Trânsito entre 1995 e 2005')
plt.grid(True)
plt.show()

```

Neste scatter plot, cada país é representado por uma bolinha. A posição de cada bolinha no gráfico é determinada pelo número total de mortes no trânsito em 1985 (eixo x) e em 2008 (eixo y). O tamanho da bolinha representa o número de mortes no trânsito em 1995, e a cor da bolinha representa a variação no número de mortes no trânsito entre 1995 e 2005.

-----

In this scatter plot, each country is represented by a dot. The position of each marble on the graph is determined by the total number of traffic deaths in 1985 (x-axis) and in 2008 (y-axis). The size of the ball represents the number of traffic deaths in 1995, and the color of the ball represents the variation in the number of traffic deaths between 1995 and 2005.
